### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df[0:50]
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
rides_per_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
rides_per_type
```

```python
# 2. Get the total drivers for each city type
drivers_per_type = city_data_df.groupby(['type']).sum()['driver_count']
drivers_per_type
```

```python
#  3. Get the total amount of fares for each city type
fare_per_type = pyber_data_df.groupby(['type']).sum()['fare']
fare_per_type
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_per_type = pyber_data_df.groupby(['type']).mean()['fare']
average_fare_per_type
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_per_driver = fare_per_type / drivers_per_type
average_fare_per_driver
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_per_driver = fare_per_type / drivers_per_type
average_fare_per_driver = pd.Series(average_fare_per_driver)
average_fare_per_driver.rename("fare")
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.merge(rides_per_type, drivers_per_type, how="left", on=['type'])
pyber_summary_df = pd.merge(pyber_summary_df, fare_per_type, how="left", on=['type'])
pyber_summary_df = pd.merge(pyber_summary_df, average_fare_per_type, how="left", on=['type'])
pyber_summary_df = pd.merge(pyber_summary_df, average_fare_per_driver.rename('fare'), how="left", on=['type'])
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None

pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df = pyber_summary_df.rename(columns={'ride_id': 'Total Rides','driver_count':'Total Drivers','fare_x':'Total Fares','fare_y':'Avgerage Fare per Ride','fare':'Avgerage Fare per Driver'})
#format for $
#pyber_summary_df = pyber_summary_df.style.format({'Total Rides':"{0:.0f}", 'Total Drivers':"{0:.0f}", 'Total Fares':"${0:,.2f}", 'Avgerage Fare per Ride':"${0:,.2f}", 'Avgerage Fare per Driver':"${0:,.2f}"})
pyber_summary_df['Total Fares ($)'] = pyber_summary_df['Total Fares ($)'].map("${0:,.2f}".format)                               
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
##date_type_df = pd.DataFrame(pyber_data_df.groupby['type']).sum()['date']
date_type = pyber_data_df.groupby(["type","date"]).sum()["fare"]
date_type

date_type_df = pd.DataFrame({"type":pyber_data_df["type"], "date":pyber_data_df["date"],"fare":pyber_data_df["fare"]})
date_type_df = pyber_data_df.groupby(["type","date"]).sum()[["fare"]]
date_type_df
#print(type(date_type_df))
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
date_type_df = date_type_df.reset_index()
date_type_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
date_type_pivot = date_type_df.pivot(index='date', columns='type', values='fare')
date_type_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
date_type_pivot = date_type_pivot.loc["2019-01-01":"2019-04-29"]
date_type_pivot
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
date_type_pivot.index = pd.to_datetime(date_type_pivot.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
date_type_pivot.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
weekly_fare_df = date_type_pivot.resample("W").sum()
weekly_fare_df
```

```python
# 9. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
fig, ax = plt.subplots(figsize=(20,7))
#ax.plot(weekly_fare_df, label = 'type')
ax.plot(weekly_fare_df["Rural"], label = 'Rural')
ax.plot(weekly_fare_df["Suburban"], label = 'Suburban')
ax.plot(weekly_fare_df["Urban"], label = 'Urban')
ax.set_ylabel('Fare ($USD)')
ax.set_title('Total Fare by City Type')
ax.legend(title= 'type', loc="center")
month_names = ['Jan 2019', 'Feb', 'Mar', 'Apr']
ax.set_xticklabels(month_names)
ax.xaxis.set_major_locator()
# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

```
